## Exploratory Data Analysis (EDA)

### Mencari Korelasi

```python
# Membuat matriks korelasi dari DataFrame df_day
correlation_matrix = df_day.corr()

# Membuat plot heatmap dengan seaborn
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")

# Menambahkan judul
plt.title("Korelasi antara Variabel Numerik")

# Menampilkan plot
plt.show()
```

### Distribusi Variabel Numerik

```python
numeric_cols = ['temp', 'atemp', 'hum', 'windspeed', 'casual', 'registered', 'cnt']

for col in numeric_cols:
    # Membuat histogram untuk setiap variabel numerik
    plt.hist(df_day[col])
    
    # Menambahkan judul
    plt.title(f'Distribusi {col}')

    # Menampilkan histogram
    plt.show()
```

### Mencari Hubungan antara musim dengan jumlah penyewa

```python
# Membuat box plot dengan seaborn
plt.figure(figsize=(10, 6))
sns.boxplot(x='season', y='cnt', data=df_day)

# Menambahkan judul
plt.title('Hubungan antara Musim dan Jumlah Sewa')

# Menampilkan plot
plt.show()
```

### Mencari Hubungan Antara Cuaca dengan Jumah Penyewa

```python
# Membuat box plot dengan seaborn
plt.figure(figsize=(10, 6))
sns.boxplot(x='weathersit', y='cnt', data=df_day)

# Menambahkan judul
plt.title('Hubungan antara Cuaca dan Jumlah Sewa')

# Menampilkan plot
plt.show()
```

### Scatter plot antara suhu dengan jumlah sewa

```python
# Membuat scatter plot dengan matplotlib
plt.figure(figsize=(10, 6))
plt.scatter(df_day['temp'], df_day['cnt'])

# Menambahkan judul dan label sumbu
plt.title('Scatter Plot antara Suhu dan Jumlah Sewa')
plt.xlabel('Suhu')
plt.ylabel('Jumlah Sewa')

# Menampilkan plot
plt.show()
```

## Visualization & Explanatory Analysis

### Pertanyaan 1: Berapa Total Jumlah Orang menyewa sepeda dalam musim panas (musim=2) pada tahun 2011

```python
# Filter tahun 2011 dan musim panas (season 2)
filtered_data = df_day[(df_day["yr"] == 0) & (df_day["season"] == 2)]

# Hitung jumlah total sewa sepeda (cnt)
total_sewa_sepeda = filtered_data["cnt"].sum()

print("Jumlah total sewa sepeda untuk tahun 2011 selama musim panas (musim 2):", total_sewa_sepeda)
```

### Pertanyaan 2: Bagaimana cara meningkatkan jumlah rentalan sepela yang digunakan oleh pengguna biasa (casual) pada hari kerja = 1

```python
# Filter pengguna casual pada hari kerja
filtered_data = df_day[(df_day["workingday"] == 1) & (df_day["casual"] > 0)]

# Membuat bar plot dengan matplotlib
plt.figure(figsize=(10, 6))
plt.bar(filtered_data['weekday'], filtered_data['casual'])

# Menambahkan judul dan label sumbu
plt.title('Jumlah Sewa Sepeda Casual pada Hari Kerja')
plt.xlabel('Hari Kerja')
plt.ylabel('Jumlah Sewa Sepeda Casual')

# Menampilkan plot
plt.show()
```

### Pertanyaan 5 : Berapa distribusi per jam sewa sepeda (cnt) pada hari natal (hari libur = 1) pada tahun 2011 (tahun =0)

```python
# Filter data tahun 2011 (yr = 0), Hari Natal (holiday = 1), dan musim panas (season 2)
filtered_data = df_hour[(df_hour["yr"] == 0) & (df_hour["holiday"] == 1) & (df_hour["season"] == 2)]

# Hitung distribusi per jam sewa sepeda (cnt)
distribusi_per_jam = filtered_data.groupby("hr")["cnt"].sum()

print("Distribusi per jam sewa sepeda pada Hari Natal tahun 2011:")
print(distribusi_per_jam)
```

Distribusi ini menunjukkan bahwa pada pukul 17:00 (5:00 sore), terjadi lonjakan paling tinggi dalam jumlah sewa sepeda dengan 1.117 sewa, yang mungkin karena banyak orang menghabiskan waktu sore untuk bersepeda saat liburan Natal. Selain itu, aktivitas sewa sepeda mulai meningkat sejak pukul 6:00 pagi dan mencapai puncaknya antara pukul 17:00 dan 18:00. Setelah itu, jumlah sewa sepeda mulai turun secara perlahan hingga larut malam. 
Ini menggambarkan pola penggunaan sepeda selama Hari Natal, yang mencerminkan kegiatan liburan dan rutinitas masyarakat pada tanggal tersebut.
